```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,234])
y = np.array([1,20])

plt.plot(x,y)
plt.show()
```

```python
# Default x points

y = np.array([1, 76, 3])
plt.plot(y)
plt.show()
```

```python
# Markers

x = np.array([1,236])
y = np.array([5,79])

plt.plot(x,y,marker='o')
plt.show()
```

```python
x = np.array([1,236])
y = np.array([5,79])

plt.plot(x,y,'o')
plt.show()
```

```python
x = np.array([1,236])
y = np.array([5,79])

plt.plot(x,y,marker='o', linestyle=':',color='r')
plt.show()
```

```python
x = np.array([1,22,33,35,42,57,67])
y = np.array([2,4,6,7,8,10,13])
plt.plot(x,y)
plt.title("test graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid()
plt.show()
```

```python
# Subplot

x = np.array([11,24,13,])
y = np.array([13,45,48])
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Sales")

x = np.array([171,224,131,])
y = np.array([153,645,848])
plt.subplot(1,2,2)
plt.title("Income")
plt.suptitle("My shop")
plt.plot(x,y)
plt.show
```

```python
# Scatter

x = np.array([21,43,45,765,34,67,215,754,35,15,21,11])
y = np.array([34,89,43,378,12,86,132,462,34,97,54,43])

plt.scatter(x,y)
plt.show()
```

```python
size = np.array([20,50,100,200,500,1000,60,90,10,300,600,45])

plt.scatter(x,y,s=size,alpha=0.5)
plt.show()
```

```python
# bargraph

x = np.array(['A','B','C','D','E'])
y = np.array([12,45,65,34,47])

plt.bar(x,y)
plt.show()
```

```python
x = np.array(['A','B','C','D','E'])
y = np.array([12,45,65,34,47])

plt.barh(x,y)
plt.show()
```

```python
x = np.array(['A','B','C','D','E'])
y = np.array([12,45,65,34,47])

plt.bar(x,y,color='hotpink',width=0.5)
plt.show()
```

```python
# Histogram

x = np.random.normal(170,10,250)

plt.hist(x)
plt.show()
```

```python
# Piechart

y = np.array([25,24,21,15])
labels=["apple","orange","grapes",'mango']

exp = [0.2,0,0,0]

plt.pie(y,labels=labels,startangle=90,explode=exp, shadow=True)
plt.legend(title="Fruits")
plt.show()
```
