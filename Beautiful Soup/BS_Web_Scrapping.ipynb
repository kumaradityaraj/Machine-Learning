```python
import requests
import pandas as pd
```

```python
page = requests.get("https://www.flipkart.com/mobiles/pr?sid=tyy%2C4io&otracker=categorytree&p%5B%5D=facets.brand%255B%255D%3DAPPLE")
page.content
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
mobiles = soup.find_all("div", class_='_13oc-S')
```

```python
print(len(mobiles))
```

```python
names=[]
prices=[]
front_camera=[]
ram=[]
for mobile in mobiles:
  name=mobile.find("div", class_="_4rR01T").text
  names.append(name)
  price=mobile.find("div", class_="_30jeq3 _1_WHN1").text
  prices.append(price)
  ul = soup.find("ul", class_='_1xgFaf')
  camera=(ul.find_all("li")[2].text)
  fc = camera.split(" | ")[1].split(" ")[0]
  front_camera.append(fc)
```

```python

dic = {'names':names,'prices':prices,'front_camera':front_camera}
df = pd.DataFrame(dic)
```
